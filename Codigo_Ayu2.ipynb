```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu=0.5474*10**-3
Cp=4181
k=0.643
Ti=25
To=75
Ts=100
dTml=((Ts-Ti)-(Ts-To))/np.log((Ts-Ti)/(Ts-To))
Pr=Cp*mu/k
```

```python
def Reynolds (F,D):
    return 4*F/mu/D/np.pi

def Nusselt (Re):
    return 0.023*Re**0.8*Pr**0.4

def calc_h (Nu,D):
    return Nu*k/D

def calc_L (h,F,D):
    l=F*Cp*(To-Ti)/(np.pi*D*dTml*h)
    return l
```

```python
diametros=np.linspace(0.03,0.05,30)
flujos=[1,2,3]
Largos=[]

for F in flujos:
    largo=[]
    Re=Reynolds(F,diametros)
    Nu=Nusselt(Re)
    h_calc=calc_h(Nu,diametros)
    L_calc=calc_L(h_calc,F,diametros)
    Largos.append(L_calc)
```

```python
plt.figure(figsize=(15,8))
plt.plot(diametros,Largos[0],'b')
plt.plot(diametros,Largos[1],'g')
plt.plot(diametros,Largos[2],'r')
plt.legend(['F=1 kg/s','F=2 kg/s','F=3 kg/s'],fontsize = 14, loc='best')
plt.xlabel('Diametro (m)', fontsize = 14)
plt.ylabel('Largo (m)',fontsize = 14)
plt.grid()
plt.savefig("Grafico_Ayu_2.jpg",bbox_inches='tight')
```
